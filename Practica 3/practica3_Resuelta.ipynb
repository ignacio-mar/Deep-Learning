d) Utilice la técnica de descenso de gradiente estocástico para obtener los coeficientes de la recta de
regresión que permita predecir el tiempo promedio que un corredor tardará en recorrer la distancia

```python
# Inicialización 
w = np.random.uniform(-1, 1)  
b = np.random.uniform(20, 35)
max_iter=1000
cota_error=0.00001
alpha=0.0005
X=horas_entrenamiento
Y=tiempo_promedio
iteraciones = 0
ecm_anterior = float('inf') 
cambio_error = cota_error* 2 
m=len(X)
# Bucle Principal de Entrenamiento (SGD)

while (iteraciones< max_iter) and (cambio_error > cota_error):
    
    #Guardamos el ECM de la época anterior para la comparación
    ecm_ant_actual = ecm_anterior 
    
    # Reiniciamos el acumulador de error total para esta Época
    suma_error_total = 0 

    #Bucle FOR: Recorre una ÉPOCA completa y actualiza W y b 
    for i in range(len(X) - 1):
        x_i = X[i]
        y_i = Y[i]
        
        # Predicción instantánea
        prediccion = w* x_i + b
        
        # ERROR RESIDUAL CUADRATICO (y - ŷ)
        error_residual = (y_i - prediccion)
        
        # ACUMULACIÓN MANUAL DEL ERROR CUADRÁTICO 
        suma_error_total += error_residual**2 
        
        # RADIENTES (Derivada de (Error^2) 
        grad_w = -2 * error_residual * x_i
        grad_b = -2 * error_residual
        
        # ACTUALIZACIÓN DE PARÁMETROS
        w = w - alpha * grad_w  
        b = b - alpha * grad_b
        
    # CÁLCULO DEL ECM TOTAL
    ecm_actual = suma_error_total / m
    
    # Cálculo de la convergencia
    cambio_error = np.abs(ecm_actual - ecm_ant_actual)
    ecm_anterior = ecm_actual 
    
    iteraciones += 1

#  Resultados Finales 
print(f"Épocas totales: {iteraciones}")
print("Coeficientes de la Recta de Regresión:")
print(f"  W (Pendiente)    : {w:.4f}")
print(f"  b (Intercepto)   : {b:.4f}")
print(f"  ECM Final        : {ecm_actual:.4f}")
print(f"Ecuación final: Tiempo ≈ {w:.4f} * Horas + {b:.4f}")
```

```python

## HECHO CON IA

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.metrics import r2_score, mean_squared_error

# -----------------------------------------------------------------
# FUNCIONES (Las tuyas, pero 'alpha' y 'm' deben ser parámetros)
# -----------------------------------------------------------------
def calcular_ecm(w, b, X_arr, Y_arr, m_val):
    return np.sum((Y_arr.flatten() - (w * X_arr.flatten() + b))**2) / m_val

def gradiente_w(w, b, X_arr, Y_arr, m_val):
    error_residual = Y_arr.flatten() - (w * X_arr.flatten() + b)
    return np.sum(-error_residual * X_arr.flatten()) / m_val

def gradiente_b(w, b, X_arr, Y_arr, m_val):
    error_residual = Y_arr.flatten() - (w * X_arr.flatten() + b)
    return np.sum(-error_residual) / m_val

def entrenamiento(X, Y, m_train, alpha, cota_error=0.0001, max_iter=10000):
    w = np.random.uniform(-1, 1)  
    b = np.random.uniform(-1, 1) # B más genérico
    
    ecm_anterior = float('inf')  
    cambio_error = cota_error * 2  
    iteraciones = 0
    
    while (iteraciones < max_iter) and (cambio_error > cota_error):
        ecm_actual = calcular_ecm(w, b, X, Y, m_train)
        grad_w = gradiente_w(w, b, X, Y, m_train)
        grad_b = gradiente_b(w, b, X, Y, m_train)
        
        w = w - alpha * grad_w
        b = b - alpha * grad_b
        
        cambio_error = np.abs(ecm_actual - ecm_anterior)
        ecm_anterior = ecm_actual
        iteraciones += 1
        
    print(f"  Entrenado en {iteraciones} iteraciones.")
    return w, b

# -----------------------------------------------------------------
# 1. CARGA Y DIVISIÓN DE DATOS
# -----------------------------------------------------------------
# m=len(X) # No definas 'm' globalmente, es local al train/test
X = np.array(df["engine-size"]).reshape(-1, 1)
Y = np.array(df["price"]).reshape(-1, 1)

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

m_train = len(X_train)
m_test = len(X_test)

```

```python
# -----------------------------------------------------------------
# 2. ENTRENAMIENTO
# -----------------------------------------------------------------

# --- i. Sin Normalizar ---
print("ENTRENANDO: SIN NORMALIZAR")
# Alpha muy bajo requerido para datos grandes
w_sin, b_sin = entrenamiento(X_train, Y_train, m_train, alpha=0.0000005) 

# --- ii. Normalización Lineal (MinMax) ---
print("ENTRENANDO: MINMAX")
# ¡Cuatro scalers distintos!
scaler_X_mm = preprocessing.MinMaxScaler()
scaler_Y_mm = preprocessing.MinMaxScaler()

# Ajustamos (fit) SOLO con datos TRAIN
X_train_mm = scaler_X_mm.fit_transform(X_train)
Y_train_mm = scaler_Y_mm.fit_transform(Y_train)

# Alpha más grande porque los datos están en [0, 1]
w_mm, b_mm = entrenamiento(X_train_mm, Y_train_mm, m_train, alpha=0.01) 

# --- iii. Estandarización (StandardScaler) ---
print("ENTRENANDO: STANDARD")
scaler_X_std = preprocessing.StandardScaler()
scaler_Y_std = preprocessing.StandardScaler()

# Ajustamos (fit) SOLO con datos TRAIN
X_train_std = scaler_X_std.fit_transform(X_train)
Y_train_std = scaler_Y_std.fit_transform(Y_train)

w_std, b_std = entrenamiento(X_train_std, Y_train_std, m_train, alpha=0.01)

# -----------------------------------------------------------------
# 3. PRUEBA Y COMPARACIÓN (La parte justa)
# -----------------------------------------------------------------
print("\n" + "="*50)
print("RESULTADOS DE PRUEBA (Comparación Justa en $)")
print("="*50)

# i. Testeo SIN NORMALIZAR
Y_pred_sin = w_sin * X_test.flatten() + b_sin
r2_sin = r2_score(Y_test, Y_pred_sin)
print(f"  R² (Sin Normalizar): {r2_sin:.4f}")

# ii. Testeo MINMAX
# Usamos .transform() (NO fit_transform) en el X_test
X_test_mm = scaler_X_mm.transform(X_test) 
Y_pred_norm_mm = w_mm * X_test_mm.flatten() + b_mm
# Desnormalizamos Y usando el scaler_Y_mm
Y_pred_mm = scaler_Y_mm.inverse_transform(Y_pred_norm_mm.reshape(-1, 1))
r2_mm = r2_score(Y_test, Y_pred_mm)
print(f"  R² (MinMax):         {r2_mm:.4f}")

# iii. Testeo STANDARD
# Usamos .transform() (NO fit_transform) en el X_test
X_test_std = scaler_X_std.transform(X_test)
Y_pred_norm_std = w_std * X_test_std.flatten() + b_std
# Desnormalizamos Y usando el scaler_Y_std
Y_pred_std = scaler_Y_std.inverse_transform(Y_pred_norm_std.reshape(-1, 1))
r2_std = r2_score(Y_test, Y_pred_std)
print(f"  R² (Standard):       {r2_std:.4f}")
print("="*50)
```

```python
import numpy as np

# Datos de Entrada (X0, X1, X2) -> [bit menos significativo (2^0), 2^1, 2^2]
# Simplificamos el problema a UNA SOLA ENTRADA X con 3 atributos: [X2, X1, X0]
X_entradas = np.array([
    [0, 0, 0],  # 0
    [0, 1, 0],  # 2
    [1, 0, 1]   # 5
])

# Salidas Esperadas (Y)
Y_salidas = np.array([0, 2, 5])

# Número de ejemplos y atributos
nEjem = X_entradas.shape[0]  # 3
nAtrib = X_entradas.shape[1] # 3

```

```python
def entrenamiento_regresion(X, Y, n_epochs, alpha, con_bias=True, w_inicial=None, b_inicial=None):
    
    m = len(X)
    n_features = X.shape[1]
    
    # Inicialización de Pesos
    if w_inicial is None:
        W = np.random.uniform(-0.5, 0.5, size=n_features)
    else:
        W = np.array(w_inicial)

    if con_bias:
        b = b_inicial if b_inicial is not None else np.random.uniform(-0.5, 0.5)
    else:
        # Anulamos el sesgo (W0 = 0)
        b = 0 
    
    # Convertir b a array si es necesario para operaciones NumPy
    b = np.array([b]) 
    
    # Variables de control
    ecm_historico = []

    # Bucle principal (Épocas)
    for ite in range(n_epochs):
        
        suma_error_total = 0.0
        
        # Recorrido Estocástico (Actualización por cada ejemplo)
        for e in range(m):
            x_e = X[e]
            y_e = Y[e]
            
            # 1. Combinador Lineal (Predicción)
            neta = np.dot(W, x_e) + b
            prediccion = neta 
            
            # 2. Error Residual
            error_residual = y_e - prediccion 
            
            # 3. Gradientes y Actualización
            # Gradiente Wj: -2 * error * Xj (multiplicación vectorial)
            grad_W = -2 * error_residual * x_e
            
            # Gradiente b: -2 * error * 1
            grad_b = -2 * error_residual * 1
            
            # Actualización
            W = W - alpha * grad_W
            if con_bias:
                b = b - alpha * grad_b
            
            suma_error_total += error_residual**2
        
        # Final de época: Calcular ECM
        ecm_actual = suma_error_total / m
        ecm_historico.append(ecm_actual)
        
        # Criterio de parada simple: si el ECM es muy bajo, detenemos
        if ecm_actual < 0.001:
            break
            
    return W, b, ite + 1, ecm_actual
    
```

```python
# Entrenar con bias (W0 es la entrada x0, b es el sesgo, que aquí llamamos b)
W_inicial_a = [4, 2, 1] # Valores ideales
b_inicial_a = 0
ALPHA_A = 0.001
MAX_ITE_A = 2000

print("--- a) ENTRENAMIENTO CON BIAS (b) ---")
W_a, b_a, ite_a, ecm_a = entrenamiento_regresion(
    X_entradas, Y_salidas, MAX_ITE_A, ALPHA_A, 
    con_bias=True, w_inicial=W_inicial_a, b_inicial=b_inicial_a
)

print(f"Iteraciones necesarias: {ite_a}")
print(f"W final (de x2, x1, x0): {W_a.round(4)}")
print(f"Sesgo (b) final: {b_a[0]:.4f}")
```

```python
# Entrenar sin bias (Forzamos b=0 y entrenamos solo W)
W_inicial_b = [0, 0, 0] # Empezamos aleatoriamente para ver qué tan rápido aprende
b_inicial_b = 0
ALPHA_B = 0.001
MAX_ITE_B = 2000

print("\n--- b) ENTRENAMIENTO SIN BIAS (b=0) ---")
W_b, b_b, ite_b, ecm_b = entrenamiento_regresion(
    X_entradas, Y_salidas, MAX_ITE_B, ALPHA_B, 
    con_bias=False, w_inicial=W_inicial_b, b_inicial=b_inicial_b
)

print(f"Iteraciones necesarias: {ite_b}")
print(f"W final (de x2, x1, x0): {W_b.round(4)}")
print(f"Sesgo (b) final: {b_b[0]:.4f}")
```

```python
# imprimir los primeros valores de clase
print("Y_train\n", Y_train[0:5])

# Crear instancia del OneHotEncoder
enc = preprocessing.OneHotEncoder(categories='auto', handle_unknown='ignore')

# Ajustar el codificador y transformar los datos
# utilizar fit_transform sobre los datos de entrenamiento y fit sobre los de testeo
Y_trainB = enc.fit_transform(Y_train.reshape(-1,1)).toarray()
Y_testB = enc.transform(Y_test.reshape(-1,1)).toarray()

# Imprimir los primeros resultados codificados
print("Y_trainB\n", Y_trainB[0:5])

# Obtener las categorías detectadas automáticamente
categorias_detectadas = enc.categories_
print("Categorías detectadas:", categorias_detectadas)
```

```python
# Opciones para la función de activación: 'sigmoid', 'tanh' o 'softmax'
# Opciones para la función de costo : 'ECM', 'EC_binaria', 'EC'
# Combinaciones válidas: 'sigmoid'-'ECM', 'sigmoid'-'EC_binaria', 'tanh'-'ECM', 'softmax'-'EC'
from grafica import *
from ClassRNMulticlase import RNMulticlase
```

```python
def entrenamiento(funcion,funcionCOSTO,Alpha,X_train,Y_trainB,Y_testB):
    if (funcion == 'tanh'):
        Y_trainB = 2*Y_trainB-1
        Y_testB = 2*Y_testB-1
        
    
    rn = RNMulticlase(alpha=Alpha, n_iter=400, cotaE=1e-06, FUN=funcion, COSTO=funcionCOSTO, random_state=None)
    result=rn.fit(X_train, Y_trainB)
    print("Pesos obtenidos\n")
    print(rn.w_)
    print(rn.b_)

```
