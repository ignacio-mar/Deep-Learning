```python
%matplotlib inline
import numpy as np
import time
#from matplotlib import pyplot as plt
from matplotlib import pylab as plt
from IPython import display
# Ajustar el tama침o predeterminado de las figuras
plt.rcParams['figure.figsize'] = [4,3] 

from grafica import *
from ClassNeuronaGral import NeuronaGradiente
```

```python
# Ejemplos de entrada de la funci칩n AND
X = np.array([[0,0], [0,1],[1,0],[1,1]])
X = 2*X-1
T = np.array([0,0,0,1])
#T=2*T-1
```

```python
ppn = NeuronaGradiente(alpha=0.15, n_iter=40, cotaE=1e-05, FUN='sigmoid', random_state=None, draw=1, title=['x1', 'x2'])
ppn.fit(X,T)
```

```python
plt.plot(range(1, len(ppn.errors_) + 1), ppn.errors_, marker='o')
plt.xlabel('Iteraciones')
plt.ylabel('ECM')
plt.show()
```

```python
Y = (ppn.predict(X)>0.5)*1
print("Y = ", Y)
print("T = ", T)
aciertos = sum(Y == T)
print("aciertos = %d   (%.2f%%)" % (aciertos, 100*aciertos/X.shape[0]))
```
