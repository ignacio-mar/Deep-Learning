```python
%matplotlib inline
import numpy as np
import pandas as pd
from matplotlib import pylab as plt
# Ajustar el tamaño predeterminado de las figuras
plt.rcParams['figure.figsize'] = [4,3]

from IPython import display

from ClassNeuronaLineal import NeuronaLineal
```

```python
Ptos = np.array([[1,1], [3,2], [5,5], [2,3], [4,3], [3.5,3], [4.3,4.2]])
X = Ptos[:,0].reshape(-1,1)
Y = Ptos[:,1]
print(X)
print(Y)
```

```python
plt.clf()
nl = NeuronaLineal(alpha=0.01, n_iter=30, cotaE=10e-03, draw=1, title=['X', 'Y'])
# --- utilice random_state=None para que los pesos se inicializacen en forma aleatoria --
nl.fit(X, Y)
```

```python
print("W = ", nl.w_)
print("b = ", nl.b_)
```

```python
plt.clf()
plt.plot(range(1, len(nl.errors_) + 1), nl.errors_, marker='o')
plt.xlabel('Iteraciones')
plt.ylabel('Cantidad de actualizaciones')
plt.show()
```
